##  Step 3: Run the Linear Regression Model



```python
from sklearn.preprocessing import StandardScaler

x = housePrices_small.iloc[:,1:].values
y = housePrices_small.iloc[:,0].values

sc = StandardScaler()
x_ = sc.fit_transform(x)

# Split the data into training/testing sets
x_train = x_[:-200]
x_test = x_[-200:]

# Split the targets into training/testing sets
y_train = y[:-200]
y_test = y[-20:0]
```

```python
import keras
from keras.layers import Input, Dense
from keras.models import Model, Sequential

inputs = Input(shape=(4,))
preds = Dense(1,activation='linear')(inputs)

model = Model(inputs=inputs,outputs=preds)
sgd=keras.optimizers.SGD()
model.compile(optimizer=sgd ,loss='mse')
model.fit(x_train,y_train, batch_size=1000, verbose=1, epochs=10, shuffle=False)
```

**=> What is the meaning of coefficients and intercept? **

```python
from sklearn.metrics import mean_squared_error, r2_score

y_test = y[:-200]
x_test = x_[:-200]

y_pred = model.predict(x_test)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, y_pred))
# Explained variance score: 1 is perfect prediction
print('r-squared score: %.2f' % r2_score(y_test, y_pred))

```
