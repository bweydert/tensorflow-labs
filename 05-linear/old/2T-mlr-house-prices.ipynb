##  Step 3: Run the Linear Regression Model



```python

x = housePrices_small.iloc[:,1:]
y = housePrices_small.iloc[:,0]

x = (x - x.mean()) / x.std() # Scale features

x
```

```python

# Split the data into training/testing sets
train_x = x[:-200]
test_x = x[-200:]

# Split the targets into training/testing sets
train_y = y[:-200]
test_y = y[-200:]
```

```python
# Specify that all features have real-value data
feature_columns = []
for key in housePrices_small.iloc[:,1:].keys():
    feature_columns.append(tf.feature_column.numeric_column(key=key))
feature_columns
```

```python
model = tf.estimator.LinearRegressor(feature_columns=feature_columns,
                                            model_dir="/tmp/lr_model")
```

```python
# Define the input function

def train_input_fn(features, labels, batch_size):
    """An input function for training"""
    # Convert the inputs to a Dataset.
    dataset = tf.data.Dataset.from_tensor_slices((dict(features), labels))

    # Shuffle, repeat, and batch the examples.
    dataset = dataset.shuffle(1000).repeat().batch(batch_size)

    # Return the dataset.
    return dataset

# Fit model.
# Train the Model.
model.train(
    input_fn=lambda:train_input_fn(train_x, train_y, BATCH_SIZE), steps=10000)
```

```python
def eval_input_fn(features, labels, batch_size):
    """An input function for evaluation or prediction"""
    
    features=dict(features)
    inputs = features if labels is None else (features, labels)
   
    # Convert the inputs to a Dataset.
    return tf.data.Dataset.from_tensor_slices(inputs).batch(batch_size)


# Evaluate the model.
eval_result = model.evaluate(
    input_fn=lambda:eval_input_fn(test_x, test_y, BATCH_SIZE))

print('\nTest set accuracy: {average_loss:0.3f}\n'.format(**eval_result))
```
